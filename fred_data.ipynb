```python
from fredapi import Fred
from dotenv import load_dotenv
import matplotlib.pyplot as plt

import os

load_dotenv()

fred = Fred(
    api_key=os.getenv('FRED_API_KEY')
)
```

```python
mortgage_rates = fred.get_series("MORTGAGE30US", observation_start="2015-01-01")
```

```python
mortgage_rates.to_frame(name="rate").to_csv("data/fred/mortgage_rates.csv")
```

```python

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(
    mortgage_rates.keys(),
    mortgage_rates.values
)
ax.set_title("30-Year Fixed Rate Mortgage Average 2015-present")
ax.set_xlabel("Date")
ax.set_ylabel("30-Year Fixed Rate Mortgage Average")
plt.show()
```
